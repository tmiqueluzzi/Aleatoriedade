```python
import xml.etree.ElementTree as ET
from openpyxl import Workbook

# Nome do arquivo XML
xml_file = '/ColeçaoRekordBox22.12.25.xml'

# Ler XML
tree = ET.parse(xml_file)
root = tree.getroot()

# Criar Excel
wb = Workbook()
ws = wb.active
ws.title = "Musicas"

# Cabeçalhos
ws.append(["Nome da Faixa", "Artista"])

# Encontrar COLLECTION
collection = root.find("COLLECTION")

# Percorrer todas as TRACK
for track in collection.findall("TRACK"):
    nome = track.attrib.get("Name", "")
    artista = track.attrib.get("Artist", "")
    ws.append([nome, artista])

# Salvar arquivo
wb.save("musicas_rekordbox.xlsx")

print("Arquivo musicas_rekordbox.xlsx criado com sucesso!")
```
